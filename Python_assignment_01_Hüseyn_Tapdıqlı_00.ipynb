# Tapşırıq 1. FizzBuzz oyunu

Bu hissədə etməli olduğunuz FizzBuzz oyununu yazmaqdır. İlk öncə oyunun şərtlərilə tanış olaq:

* Müəyyən bir ədəd seçilir və 1dən başlayaraq bu ədəd doğru sayılır
* Əgər sayılan ədəd:
    * 3ə bölünürsə həmin ədədin əvəzinə "Fizz";
    * 5ə bölünürsə həmin ədədin əvəzinə "Buzz";
    * Həm 3ə həm 5ə bölünürsə həmin ədədin əvəzinə "FizzBuzz" yazılır
    * Bu şərtlərin heç biri ödənmədikdə sadəcə ədədin özü deyilir

--- 

Bu tapşırıq üçün FizzBuzz adlı bir funksiya yazmalısınız. Bu funksiya istifaddəçidən input olaraq limit ədədi soruşur və bu ədədə qədər saymağa başlayaraq yuxarıdakı qaydalara uyğun olaraq nəticə çıxarmalıdır. Bundan əlavə funksiya sonda həmin ədədə qədər olan ümumi Fizz, Buzz və FizzBuzz-ların sayını çap etməlidir.

```python
# FizzBuzz oyunu

def func_fizzbuzz():
  print("FizzBuzz oyununa xoş gəlmisiniz! :)\n")
  # typea, typeb,typec --burada her kateqoriyadan nece eddedir 0-in ustune gelerek cemleyir.
  typea= 0 
  typeb= 0
  typec= 0
  #res=0 sadece while loop-nu basladmaq ucundur.
  res=0
  while res==0:
    n= int(input("Rəqəmi daxil edin: "))
    
    for a in range(1, n+1, 1):    # yazilan eded daxil olmaqla tek-tek ededleri sayir.

      '''
      5-e ve 3-e bolunen ededler. Ilk 5-e ve 3-e bolunen kateqoriyasini yerlesdirdimki, eger diger kateqoriyalari ilk verse idim. 
      Elave olaraq digerlerine a%3==0 and a%3!=0 yazmali idim.
      '''
      if a%5==0 and a%3==0: 
        print("FizzBuzz")
        typea+=1
      elif a%5==0: # uygun olaraq 5-e bolunenler
        print("Buzz")
        typeb+=1
      elif a%3==0: #3-e bolunenler
        print("Fizz")
        typec+=1
      else :       # diger ededler ucun
        print(a)
    res=1      # while dongusunu dayandirmaq ucun
  '''
  print fizz, buzz, fizzbuzz kateqoriyalarinin sayini verir
  '''      
  print(f"\nFizz sayı:     {typec}\nBuzz sayı:     {typeb}\nFizzBuzz sayı: {typea}\n------- Game Over --------")
```

```python
func_fizzbuzz()
```

# Tapşırıq 2. Fibonaçi Seriyası

Bu tapşırıqda fibonaçi seriyasını çap edən bir funksiya yazacaqsınız. Fibonaçi seriyası 0, 1 olaraq başlayır və sonrakı hər bir ədəd özündən əvvəlki 2 ədədin cəminə bərabərdir:

`0, 1, 1, 2, 3, 5, 8, 13, ...`

* 1 = 0 + 1
* 2 = 1 + 1
* 3 = 2 + 1 və s.

Fibonaçi funksiyası 2 üsulla həll olunmalıdır:
* loop-lardan istifadə ilə
* recursion-dan istifadə ilə

Funksiya istifadəçidən neçə fibonaçi ədədi çıxarmaq istədiyini soruşmlaıdır və bu sayda ardıcıllıqla fibonaçi ədədlərini çıxarmaq lazımdır.

```python
def say():
  n = int(input("dcfvc: ")) #nece eded coxarmaq lazimdi say yaziriq 
  series = [0, 1] # burada uzerine gelsin deye list yaradiriq

  for x in range(2, n): #n-ci edede qeder sayir
      next = series[x - 1] + series[x - 2]  # listdeki yerlerine gore eded cixarir
      series.append(next)  #ve cemleyir uzerine
      
  for a in series:  # listdeki ededleri sayir.
    print(a)


#bu ededlerin yaninda olan cemlemeyi bilmedim. cunki indexi yoxdur basa dusmur hansi ededle cemleyecek. normalda f"{a} = {a-1} + {a-2}" yazardim.
#Amma bu ededler mueyyen ardicilliqla gedir.
```

```python
say()
```

# Tapşırıq 3. String-də character sayı

Bu tapşırıqda yazacağınız funksiya input olaraq bir string götürməli və həmin string-dəki hərf və rəqəmlərin sayını göstərən bir dictionary qaytarmalıdır. Nümunə olaraq aşağıya baxın 

image.png

```python
#cevirici
def hist(): 
  word_input = input("Sozu daxil edin: ")  
  word_list = []  #liste sozu yigmaq ucun

  for i in range(len(word_input)):     #sozun uzunluguna gore bolur ve sirasi ile liste elave edir. 
      word_list.append(word_input[i])
  
  print(" output :", dict(map(lambda x: (x, word_list.count(x)), word_list)))  # sayina gore dictionary-lere value elave edir. ilk lambda eyni herfleri cemleyir ve list edir, sonra map keyleri ve listi birlesdirerek dicte oturur.
```

```python
hist()
```
